```python
#[문제59] 입력값을 더하는 함수를 작성하세요.
'''
>>> print(add(2))
2

>>> print(add(8))
10
'''
x = 0
def add(arg):
  global x
  x = x + arg
  return(x)

print(add(2))
print(add(8))
```

```python
# 스택(stack)
'''
- 데이터를 임시로 저장할 때 사용하는 자료형
- LIFO(Last In First Out) : 가장 나중에 입력한 데이터를 가장 먼저 꺼낸다(삭제)
- push, pop
'''
'''
[문제60] stack 변수를 선언한 후 push 함수를 이용해서 값을 입력하고 full함수를 통해 stack변수를 확인하고
pop변수를 이용해서 stack변수에 입력된 값을 꺼내는 함수를 생성해주세요.

full()        
stack 변수가 비어 있습니다.

push(100)    
0 인덱스에 값 100 입력 

push(200)    
1 인덱스에 값 200 입력 

full()  
0 인덱스 값 100
1 인덱스 값 200

v_pop = pop()
1 인덱스에 값 200 꺼냄  # 꺼낸 값을 v_pop에 저장
'''

stack = []
v_pop = []

def full():
  global stack
  if len(stack) == 0:
    print('stack 변수가 비어 있습니다')
  else:
    cnt = 0
    for i in stack:
      print(f'{cnt} 인덱스 값 {i}')
      cnt += 1

def push(arg1):
  global stack
  stack.append(arg1)
  print(f'{len(stack)-1} 인덱스에 값 {arg1} 입력')

def pop():
  global stack, v_pop
  temp = stack.pop()
  print(f'{len(stack)} 인덱스에 값 {temp} 꺼냄')
  v_pop.append(temp)


full()
push(100)
push(200)
push(300)
full()
pop()
pop()
v_pop


stack = []
def push(arg):
  stack.append(arg)
  print("{} 인덱스에 값 {} 입력".format(len(stack)-1,arg))

push(100)
push(200)

def full():
  if len(stack) ==0:
    print("stack 변수가 비어 있습니다.")
  else: 
    v_cnt = 0
    for i in stack:
      print(f"{v_cnt} 인덱스 값 {i}")
      v_cnt += 1

full()

# enumerate() : 리슽, 튜플에서 인덱스, 값을 반환하는 함수
def full():
  if len(stack) ==0:
    print("stack 변수가 비어 있습니다.")
  else: 
    for idx,value in enumerate(stack): # 인덱스와 값을 같이 출력할 수 있는 함수 enumerate
      print(f"{idx} 인덱스 값 {value}")

full()

def pop():
  if len(stack) == 0:
    print("stack 변수가 비어 있습니다.")
  else:
    v_cnt = len(stack)-1
    v_pop = stack[v_cnt]
    print(f'{v_cnt} 인덱스에 값 {v_pop} 꺼냄')
    del stack[v_cnt]
    return v_pop

full()
pop()
full()
pop()
full()
pop()
```

```python
# Queue(큐)
'''
- FIFO(First In First Out) : 가장 먼저 입력한 데이터를 가장 먼저 꺼내는 작업, 선입선출
- 입력 enqueue, 꺼내는 작업 dequeue
'''
'''
[문제61] queue 변수를 선언한 후 enqueue 함수를 이용해서 값을 입력하고 full함수를 통해 queue변수를 확인하고
dequeue변수를 이용해서 queue변수에 입력된 값을 꺼내는 함수를 생성해주세요.

'''
queue = []

def enqueue(arg):
  queue.append(arg)
  print(f'{len(queue)-1} 인덱스에 {arg} 입력')


def full():
  if len(queue) == 0:
    print('queue변수에 값이 없습니다')
  else:
    for idx,value in enumerate(queue):
      print(f'{idx} 인덱스 값 {value}')

full()
enqueue(100)
enqueue(200)
full()

def dequeue():
  if len(queue) ==0:
    print("queue 변수가 비어 있습니다.")
  else:
    v_queue = queue[0]
    del queue[0]
    print(f"0 인덱스에 값{v_queue} 꺼냄")
    
    return v_queue

dequeue()
full()
dequeue()
full()

```

```python
#[문제62] 함수를 생성하세요.
'''
합 = 관측값의 합
평균 = 관측값의 합 / 관측값의 수
편차 = 관측값 - 평균
편차제곱합 = 편차**2 + 편차**2
분산 = 편차제곱합/(관측값의 수-1(자유도))
표준편차 = math.sqrt(분산)

sum(1,2,3,4,5)
mean(1,2,3,4,5)
variance(1,2,3,4,5)
stddev(1,2,3,4,5)
'''

def sum(*arg):
  hap = 0
  for i in arg:
    hap += i
  return(hap)

def mean(*arg):
  return(sum(*arg)/len(arg))

def variance(*arg):
  total = 0
  avg = mean(*arg)
  for i in arg:
    total += (i-avg)**2
  return total/(len(arg)-1)

import math
def stddev(*arg):
  return(math.sqrt(variance(*arg)))

if __name__ == '__main__':
  print(sum(1,2,3,4,5))
  print(mean(1,2,3,4,5))
  print(variance(1,2,3,4,5))
  print(stddev(1,2,3,4,5))

```
