```python
2022.5.10
cnn

from tensorflow.keras import datasets

(x_train,y_train),(x_test,y_test) = datasets.mnist.load_data()

x_train.shape # 60000,28,28
y_train.shape
x_test.shape # 10000,28,28
y_test.shape

# grayscale
# 0(검정색) ~ 255(하얀색)
x_train[0].shape
28*28
y_train[0]

import matplotlib.pyplot as plt
plt.imshow(x_train[1000],'gray') # 숫자 확인
plt.title(y_train[1000])
plt.show()

# 정답레이블을 one hot encoding으로 변환
from tensorflow.keras.utils import to_categorical
to_categorical(0,10) # 0번째에 1 나머지는 0으로 채워진 배열(one hot encoding)
to_categorical(5,10)

train_ohe = to_categorical(y_train) # 정답 레이블 one hot encoding
test_ohe = to_categorical(y_test)
train_ohe.shape
test_ohe.shape

from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Sequential
import tensorflow as tf

model = Sequential([ # 모델 만들어 놓기
    Flatten(input_shape=(28,28)), # Flatten: 인풋
    Dense(100,activation='relu'), # 첫 번째 은닉층
    Dense(30,activation='relu'), # 두 번째 은닉층
    Dense(10,activation='softmax')]) # 출력층

model.summary()

model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
history = model.fit(x_train,train_ohe,epochs=20,batch_size=32) # 모델에 적용, history에 저장

history.history['loss'] # 모델을 수행한 history로 loss값 확인
history.history['accuracy'] # accuracy 확인

model.evaluate(x_test,test_ohe,batch_size=128) # 테스트 데이터 평가



image(입력) -> feature Extractor(특징추출),cnn layer -> Full connected layer(분류) -> 정답

★Convolutional Neural Network(합성곱)
- computer vision에서 많이 사용되는 딥러닝 모뎅
- 컴퓨터 비전은 시각적 세계를 해석하고 이해하도록 컴퓨터를 학습시키는 인공지능 분야
- 이미지 인식과 음성인식, 텍스트 분석 등 다양한 곳에서 사용되는 딥러닝 모델

convolution -> relu -> pooling -> fully connected

# convolution layer
- 이미지의 특징을 찾는 층
- feature map을 만들고 그 feature map을 선명하게 하는 층

# 합성곱 연산
- 이미지 3차원(세로,가로,색상(채널)) 데이터의 형상을 유지하면서 연산하는 작업, grayscale 1, RGB 3
- 입력 데이터에 필터를 적용한 것이 합성곱 연산
- 이미지에 대한 특성을 감지 하기 위해 커널(kernel)이 필터(filter)이다

이미지           필터
1 2 3 0         2 0 1
0 1 2 3    *    0 1 2  = 15 () 
3 0 1 2         1 0 2    () ()
2 3 0 1
입력(4*4)       필터(3*3) 출력(2*2)

1*2+2*0+3*1+0*0+1*1+2*2+3*1+0*0+2*1

# stride
- 합성곱 연산을 하기 위해서 필터를 적용하는 위치의 간격

# padding
- 합성곱 연산을 수행하기 전에 입력 데이터 주변을 특정한 값으로 채워 늘리는 작업
- 패딩을 하지 않을 경우 데이터의 공간 크기는 합성곱 계층을 지날 때 마다 작아지게 되므로
 가장 자리 정보들이 사라지게 되는 문제가 발생하기 때문에 이를 방지하고자 패딩을 사용
 
입력(4,4), 필터(3,3), 패딩 0, 스트라이드 1 -> 출력(2,2)
입력(4,4), 필터(3,3), 패딩 1, 스트라이드 1 -> 출력(4,4)
입력(h,w), 필터크기(fh,fw), 패딩(p), 스트라이드(s) -> 출력(oh,ow)

        h + 2*p - fh
oh = --------------- + 1
            s

        w + 2*p - fw
ow = --------------- + 1
            s

# pooling
- 출력값에서 일부분만 취하는 가정
- 주변의 픽셀을 묶어서 하나의 대표 픽셀로 바꾼다.
- 이미지 차원을 축소

풀링(pooling)의 종류
- 최대 풀링(pooling) : 컨블루션(convolution) 데이터에서 가장 큰값을 대표값으로 선정
- 평균 풀링(pooling) : 컨블루션(convolution) 데이터에서 모든 값의 평균을 대표값으로 선정



# sequential model 생성

model = Sequential([ # 모델 만들어 놓기
    Flatten(input_shape=(28,28)), # Flatten: 인풋
    Dense(100,activation='relu'), # 첫 번째 은닉층
    Dense(30,activation='relu'), # 두 번째 은닉층
    Dense(10,activation='softmax')]) # 출력층

model.summary()

model = Sequential() 
model.add(Flatten(input_shape=(28,28))) # Flatten: 인풋
model.add(Dense(100,activation='relu')) # 첫 번째 은닉층
model.add(Dense(30,activation='relu')) # 두 번째 은닉층
model.add(Dense(10,activation='softmax')) # 출력층
model.summary()

# functional API
from  tensorflow.keras.layers import Input,Flatten,Dense
from  tensorflow.keras.models import Model

input_size = 28

input_tensor = Input(shape=(input_size,input_size)) # 입력값의 이미지크기
x = Flatten()(input_tensor) # 입력
x = Dense(100,activation='relu')(x) # 파이프라인처럼 x에 입력
x = Dense(30,activation='relu')(x)
output = Dense(10,activation='softmax')(x)
model = Model(inputs=input_tensor,outputs=output)
model.summary()


image = x_train[0]
image.shape
label = y_train[0]
label

plt.imshow(image,'gray')
plt.title(label)
plt.show()

tf.keras.layers.ConV2D(filters=3,kernel_size(3,3),strides=(1,1),padding='same', activation='relu') #same 은 padding을 인풋아웃풋을 동일하게 하겠다
tf.keras.layers.ConV2D(filters=3,kernel_size(3,3),strides=(1,1),padding='valid') # valid는 padding을하지 않겠다
tf.keras.layers.ConV2D(3,3,1,'same')

x_train.shape #(60000, 28, 28) -> (60000, 28, 28,1)
# 28 * 28 -> (batchsize,height,weight,channel) (1, 28, 28,1)
image.shape

# 차원 수 늘리기
image = image[tf.newaxis,...,tf.newaxis]
image.shape

image.dtype
image = tf.cast(image,dtype=tf.float32)
image.dtype

import numpy as np
np.min(image)
np.max(image)

# 합성곱
layer = tf.keras.layers.Conv2D(3,3,1,'same')
output = layer(image)
output
np.min(output)
np.max(output)

plt.subplot(1,2,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,2,2)
plt.imshow(output[0,:,:,2],'gray') # 합성곱 연산을 수행한 이미지
plt.show()

filter 에 있는 값은 기존의 weight 값과 같다

w_0 = layer.get_weights()[0] # filter  filter를 찾는 것이 중요
w_0.shape # (height,weight,channel,filters)

# bias
layer.get_weights()[1] #bias

plt.subplot(1,3,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,3,2)
plt.imshow(w_0[:,:,0,0],'gray') # filter 이미지 첫번째 필터
plt.subplot(1,3,3)
plt.imshow(output[0,:,:,0],'gray') # 합성곱 연산을 수행한 이미지
plt.show()


plt.subplot(1,3,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,3,2)
plt.imshow(w_0[:,:,0,1],'gray') # filter 이미지 두번째 필터
plt.subplot(1,3,3)
plt.imshow(output[0,:,:,0],'gray') # 합성곱 연산을 수행한 이미지
plt.show()

plt.subplot(1,3,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,3,2)
plt.imshow(w_0[:,:,0,2],'gray') # filter 이미지 세번째 필터
plt.subplot(1,3,3)
plt.imshow(output[0,:,:,0],'gray') # 합성곱 연산을 수행한 이미지
plt.show()


np.min(image)
np.max(image)
# 합성곱
layer = tf.keras.layers.Conv2D(3,3,1,'same')
output = layer(image)
output
np.min(output)
np.max(output)

#Relu
layer = tf.keras.layers.ReLU()
output = layer(output)
np.min(output)
np.max(output)

plt.subplot(1,2,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,2,2)
plt.imshow(output[0,:,:,0],'gray') # 합성곱 연산을 수행한 이미지
plt.show()

# pooling
layer = tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2),padding='same')
output = layer(output)
output

plt.subplot(1,2,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,2,2)
plt.imshow(output[0,:,:,0],'gray') # 합성곱 연산을 수행한 이미지
plt.show()

# Flatten
layer = tf.keras.layers.Flatten()
output = layer(output)
output

plt.subplot(1,2,1)
plt.imshow(image[0,:,:,0],'gray') # 원본 이미지
plt.subplot(1,2,2)
plt.imshow(output[:,:100],'gray') # flatten 이미지
plt.show()

# Dense
layer = tf.keras.layers.Dense(32,activation='relu')
output = layer(output)
output.shape


#---------------------------------------------------------------
import glob # 여러 파일 읽어오는 라이브러리
from PIL import Image
#glob.glob('c:/data/101_ObjectCategories/chair/*.jpg')
#glob.glob('c:/data/101_ObjectCategories/camera/*.jpg')

caltech_dir = 'c:/data/101_ObjectCategories/'
categories = ['chair','camera','butterfly','elephant','flamingo']

'chair' -> [1,0,0,0,0]
'camera' -> [0,1,0,0,0]

# 각 이미지를 array로 변환하는 작업
x = []
y = []
image_w = 64
image_h = 64
for idx,cate in enumerate(categories):
    label = [0 for i in range(5)]
    label[idx] = 1
    image_dir = caltech_dir + cate
    files = glob.glob(image_dir + '/*.jpg')
    for i in files:
        img = Image.open(i)
        img = img.convert('RGB') # rgb로 바꾸기
        img = img.resize((image_w,image_h)) # 크기 변환 
        data = np.asarray(img)
        x.append(data) # 이미지 array로 변환
        y.append(label)

len(x)
len(y)

x = np.array(x)
x.shape # (334, 64, 64, 3) 전체 이미지가 334, 64,64 채널은3
y = np.array(y)
y.shape

x[0]
y[0]
x[333]
y[333]

plt.imshow(x[1][:,:,0])
plt.title(y[1])
plt.imshow(x[333][:,:,0])
plt.title(y[333])

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y)

x_train.shape
y_train.shape
x_test.shape
y_test.shape
xy = (x_train,x_test,y_train,y_test)
np.save('c:/data/101_ObjectCategories/my_image.npy',xy) # 저장하기

train_x,test_x,train_y,test_y = np.load('c:/data/101_ObjectCategories/my_image.npy',allow_pickle=True) # 읽어오기
train_x.shape

# 이미지 학습시에 데이터에 대한 표준화 작업을 수행하는게 학습효과가 좋다. (이미지는 보통 float형식으로 만들고 255로 나눈다)
x_train = x_train.astype('float')/255
x_test = x_test.astype('float')/255
x_train[0]
x_test[0]
x_train.shape

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPool2D, Flatten, Dense, Dropout # Dropout은 오버피팅을 줄이는데 사용

model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),strides=(1,1),padding='same',input_shape=(64,64,3),activation='relu')) 
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),padding='same',input_shape=(64,64,3),activation='relu')) 
model.add(MaxPool2D(pool_size=(2,2))) # 꼭 해야하는 것은 아님?
model.add(Conv2D(filters=64,kernel_size=(3,3),strides=(1,1),padding='same',input_shape=(64,64,3),activation='relu')) 
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.5)) # 50% 정도 없애자

model.add(Dense(5,activation='softmax'))
model.summary()

model.compile(optimizer=tf.optimizers.Adam(learning_rate=0.001),loss='categorical_crossentropy',metrics=['accuracy'])
history = model.fit(x_train,y_train,epochs=20,batch_size=32) 

model.evaluate(x_test,y_test,batch_size=64) # 테스트 데이터 적용


# 새로운 이미지를 우리가 만든 모델을 이용해서 예측 분류해보자
test = []
new = Image.open("C:/data/101_ObjectCategories/chair/image_0061.jpg")
new = new.convert('RGB') # rgb로 바꾸기
new = new.resize((64,64)) # 크기 변환 
new = np.asarray(new)
test.append(new)
new_data = np.array(test)
np.argmax(model.predict(new_data.astype('float')/255)) # 예측 



```
