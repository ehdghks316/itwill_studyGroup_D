```python
★ folium
- 지도 시각화 라이브러리
https://python-visualization.github.io/folium/
(base) c:/Users/user> pip install folium # anaconda prompt창에서 관리자권한으로 실행하고

import folium

m = folium.Map(location = [37.49760733241009, 127.02732082092184],zoom_start=13, tiles='Stamen Terrain') # 위도경도를 이용해서 맵을 가져오기
folium.Marker(location=[37.49882309502913, 127.03163884047372],
              popup='itwill',
              icon=folium.Icon(color='red',icon='star')).add_to(m) # 마커 표시 함수 
m.save("c:/data/map1.html") # 맵을 html형식으로 저장
 

m = folium.Map(location = [37.49760733241009, 127.02732082092184],zoom_start=13, tiles='Stamen Terrain') # 위도경도를 이용해서 맵을 가져오기
folium.CircleMarker(location=[37.49882309502913, 127.03163884047372],
              color='red',
              radius=30,
              tooltip='학원주변').add_to(m) # 마커 표시 함수 
m.save("c:/data/map1.html") # 맵을 html형식으로 저장

import pandas as pd
covid = pd.read_csv('c:/data/covid_19.csv')
covid.info()
covid.head()

covid.Date = pd.to_datetime(covid.Date)
lately_covid = covid[covid.Date ==covid.Date.max()]
m = folium.Map(location=[0, 0],zoom_start=13)
for i in lately_covid.index:
    folium.CircleMarker([lately_covid['Lat'][i],lately_covid['Long'][i]],
                        radius=int(lately_covid['Confirmed'][i]/100000),
                        color='red',
                        fill='True',
                        pipup=lately_covid['Country/Region'][i],
                        tooltip=int(lately_covid['Confirmed'][i])).add_to(m)
  
m.save("c:/data/map1.html") # 맵을 html형식으로 저장

[내가 분석하고자 하는 문제] 건강원들만 지도에 표시하라, 
data = pd.read_csv('c:/data/소상공인시장진흥공단_상가(상권)정보_서울_202112.csv')
data.info()
data.head()

cure = data[data.상권업종소분류명=='건강원']
m = folium.Map(location=[37.53172446135554, 127.05851199961289],zoom_start=18)
for i in cure.index:
    folium.Marker(location=[cure['위도'][i],cure['경도'][i]],
                 popup=cure['상호명'][i],
                 icon=folium.Icon(color='red',icon='star')).add_to(m)
m.save("c:/data/상권1.html")

[▶내가 분석하고자 하는 문제] 고층(10층이상?)에 식당이 있는 곳은 주로 어디일까? 그래프, 지도 표시# ( 식당이 고층에 있으면 가격대가 조금 비싸고 부유계층이 찾을 가능성이 높다고 주관적인 판단, 데이트코스로 높은 건물에서 창밖을 바라보며 고급스러운 식사)
data = pd.read_csv('c:/data/소상공인시장진흥공단_상가(상권)정보_서울_202112.csv')

#구별로 식당개수 추출하고 시각화
rest = data.groupby(data.시군구명)['상권업종대분류명'].count()
rest.plot(kind='bar') 
# 그래프를 확인해보면 압도적으로 강남이 많은 것을 알 수 있다.

# 10층이상의 상가들의 정보 추출 ( 층정보는 float형식이고 안에 데이터는 숫자, 문자+숫자, 한글 등 전처리작업을 필요로 한다. )
import re
c = data[~data.층정보.isna()]
x = c.reset_index(drop=True)

a = []
for i in c.index:
    if pd.to_numeric(re.findall('\d+',c.층정보[i])).any() != None:
        a.append(pd.to_numeric(re.findall('\d+',c.층정보[i])).any())
c1 = x[a]

c1.층정보[0].isdigit() # isdigit는 전체 인덱스에 적용하려면 for문 사용, isdigit함수는 문자열에서 숫자만 있는지 체크하는 함수
b = []
for i in c1.index:
    if c1.층정보[i].isdigit():
        b.append(pd.to_numeric(c1.층정보[i]) > 10)
    else:
        b.append(False)

cheng = c1[b]

# 10층이상의 상가들 중 식당 및 카페
data.info()
data[data.상권업종대분류명 == '음식']

result = cheng[cheng.상권업종대분류명 =='음식']

# 지도 시각화
m = folium.Map(location=[37.53172446135554, 127.05851199961289],zoom_start=13)
for i in result.index:
    folium.Marker(location=[result['위도'][i],result['경도'][i]],
                  popup=result['상호명'][i],
                  icon=folium.Icon(color='red',icon='circle')).add_to(m)
m.save('c:/data/상권2.html')

# 그래프
import matplotlib.pylab as plt
from matplotlib import font_manager,rc
font_name = font_manager.FontProperties(fname="c:/windows/fonts/HMKMMAG.TTF").get_name()
rc('font',family=font_name)

result.info()
rg = result.상호명.groupby(result.시군구명).count()

rg.plot(kind='bar')
rg.plot(kind='barh')

ax = plt.bar(rg.index,rg)
plt.xticks(rg.index, rotation=90)
plt.title('서울특별시 구별 고층 식당')
plt.ylabel('식당개수')
plt.xlabel('서울특별시')
for i in ax.patches:
    left,bottom,width,height = i.get_bbox().bounds
    plt.text(left,height,int(height))
    
[내가 분석하고자 하는 문제] 호텔은 어디에 많이 있을까? 여관, 모텔은 어디에 많이 있을까?
data.info()
data

[내가 분석하고자 하는 문제] 스타벅스는 어디에 많이 있을까? 이디야는 어디에 많이 있을까? 그 차이는? 비교분석해보기

[내가 분석하고자 하는 문제] 술집이 많이 모여있는 곳과 왜 거기에 많이 모여있을까 분석
data.info()

sul = data[data.상권업종중분류명 =='유흥주점']
cnt = sul.groupby(sul.시군구명)['상호명'].count()
cnt.plot(kind='bar')
m = folium.Map(location=[37.53172446135554, 127.05851199961289],zoom_start=13)
for i in sul.index:
    folium.CircleMarker([sul.위도[i],sul.경도[i]],
                       radius=10,
                       color='green',
                       fill=True).add_to(m)
```
