```python
# ★ feature scaling
서로 다른 변수의 값 범위를 일정한 수준으로 맞추는 작업

1. 표준화(standardization)
- 데이터의 피처 각각이 평균이 0이고 분산이 1인 정규분포를 가진 값으로 변환
- z-score
(관측값 - 관측값의 평균) / (관측값의 표준편차)
```

```python
#[문제63] standardization(표준화) 함수를 생성해주세요.
data = (100,2,4,3,500)

def sum(*arg): # 합
  total = 0
  for i in arg:
    total += i
  return total

def mean(*arg): # 평균
  return(sum(*arg)/len(arg))

def variance(*arg): # 분산
  total = 0
  avg = mean(*arg)
  for i in arg:
    total += (i-avg)**2
  return total/(len(arg)-1)

import math
def stddev(*arg): # 표준편차
  return(math.sqrt(variance(*arg)))

def standardscaler(*arg): # z-score(표준화)
  z_score = []
  v_mean = mean(*arg)
  v_stddev = stddev(*arg)
  for i in arg:
    z_score.append((i - v_mean) / v_stddev)
  return z_score

data_scaler = standardscaler(*data) # z-score
mean(*data_scaler) # 평균이 0에 가까움
stddev(*data_scaler) # 표준편차1
```
