```python
#[문제42] 과일의 빈도수를 만들어 주세요.
fruit = ("사과","귤","오렌지","배","포도","바나나","키위","딸기","블루베리", "망고",
         "수박","사과","귤","키위","포도","바나나","사과","딸기","블루베리", "망고",
         "사과","귤","오렌지","배","포도","바나나","사과","딸기","파인애플")
apple=0
watermelon=0
ggyul=0
orange=0
graph=0
banana=0
kiwi=0
strawberry=0
fineapple=0
mango=0
blueberry=0
bae=0
for i in fruit:
  if i == '사과':
    apple += 1
  elif i == '수박':
    watermelon += 1
  elif i == '귤':
    ggyul += 1
  elif i == '오렌지':
    orange += 1
  elif i == '포도':
    graph += 1
  elif i == '바나나':
    banana += 1
  elif i == '키위':
    kiwi += 1
  elif i == '딸기':
    strawberry += 1
  elif i == '파인애플':
    fineapple += 1
  elif i == '망고':
    mango += 1
  elif i == '블루베리':
    blueberry += 1
  else :
    bae += 1

print(f'사과 : {apple} 수박 : {watermelon} 귤 : {ggyul} 오렌지 : {orange} 포도 : {graph} 바나나 : {banana} 키위 : {kiwi} 딸기 : {strawberry} 파인애플 : {fineapple} 망고 : {mango} 블루베리 : {blueberry}')
print('사과 :',apple,'\n'
      '수박 :',watermelon,'\n',
      '귤 :',ggyul,'\n',
      '오렌지 :',orange,'\n',
      '포도 :',graph,'\n',
      '바나나 :',banana,'\n',
      '키위 :',kiwi,'\n',
      '딸기 :',strawberry,'\n',
      '파인애플 :',fineapple,'\n',
      '망고 :',mango,'\n',
      '블루베리 :',blueberry)

##########################################강사님 답
fruit.count("사과")
fruit.count("귤")

set(fruit) # 유니크, 유일한 값만 뽑는 함수(집합)
for i in set(fruit):
  print('{} : {}'.format(i,fruit.count(i)))

result = ''
for i in set(fruit):
  result = result + '{} : {}'.format(i,fruit.count(i)) + '\n'

print(result)

result = []
for i in set(fruit):
  result.append((i,fruit.count(i)))
result
```
