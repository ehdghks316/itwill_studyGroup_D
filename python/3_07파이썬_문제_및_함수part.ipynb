```python
#[문제44] data 변수에 있는 문장에서 공백 문자를 기준으로 분리한 후 빈도수를 구하세요. 단 딕셔러니 자료형을 이용하세요.
data = 	"별 하나에 추억과 별 하나에 사랑과 별 하나에 쓸쓸함과 별 하나에 동경과 별 하나에 시와 별 하나에 어머니 어머니"

data.split(' ')
poem = {}

for i in data.split(' '):
  if i not in poem.keys():
    poem[i] = 1
  else:
    poem[i] = poem[i] + 1

poem

words = data.split(' ')
freq_dict = {}
for i in words:
  if i not in freq_dict.keys():
    freq_dict[i] = 1 
  else:
    freq_dict[i] = freq_dict[i] + 1

sorted(freq_dict) # value값이 안나옴
sorted(freq_dict.keys())
sorted(freq_dict.values())
sorted(freq_dict.items()) # key, value 모두나옴 key값을 기준으로 정렬

for k, v in sorted(freq_dict.items()):
  print(k,v)

for k, v in sorted(freq_dict.items(), reverse = True):
  print(k,v)

import operator

# 0번인덱스는 key/ key를 기준으로 정렬
for k, v in sorted(freq_dict.items(),key=operator.itemgetter(0)): 
  print(k,v)

# 1번인덱스는 value/ value값을 기준으로 정렬
for k, v in sorted(freq_dict.items(),key=operator.itemgetter(1)): 
  print(k,v)

# key를 기준으로 내림차순정렬
for k, v in sorted(freq_dict.items(),key=operator.itemgetter(0),reverse=True): 
  print(k,v)

# value를 기준으로 내림차순정렬
for k, v in sorted(freq_dict.items(),key=operator.itemgetter(1),reverse=True): 
  print(k,v)


import collections
words_dict = collections.Counter(words) # 위의 for문 대신할 함수, 리스트 안에 단어들을 넣어 놓고 이 함수를 사용하면 dict 형태로 출력해준다.
words_dict.keys()
words_dict.values()
words_dict.items()
for k, v in sorted(words_dict.items(),key=operator.itemgetter(1),reverse=True): 
  print(k,v)
```
