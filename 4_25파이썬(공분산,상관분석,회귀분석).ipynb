```python
★ 분산(variance)
- 내가 가진 자료(데이터)가 평균값을 중심으로 퍼져있는 평균적인 거리(정도)

        합((개별관측값 - 개별평균)**2)
분산 = ----------------------------
        관측값의 수 - 1(자유도)

★ 공분산(covariance)
- 두 변수가 얼마나 함께 변하는지를 측정하는 지표
- cov(x,y) > 0 : x와 y의 변화가 같은 방향으로 변화가 된다.
                한 변수가 커질 때 다른 변수가 함께 커지거나 한 변수가 작아질 때 다른 변수가  함께 작아지는 경우 변화의 방향이 같다.

- cov(x,y) < 0 : x와 y의 변화가 다른 방향으로 변화가 된다.
                한 변수가 커질 때 다른 변수는 작아지거나 한 변수가 작아질 때 다른 변수는 커지는 경우는 변화의 방향이 다르다는 의미

- cov(x,y) = 0 : x와 y의 값이 서로 상관없이 움직일 경우 공분산이 0이다.


                합((x - x평균)*(y - y평균))
공분산(x,y) = ----------------------------
                관측값의 수 - 1(자유도)

★ 상관분석(correlation)
- 공분산을 표준화하는 방법
- 변수들 간의 연관성을 파악하기 위해 사용하는 분석기법중 하나로 변수간의 선형관계 정도를 분석하는 통계기법
- 두 변수 사이의 관련성을 파악하는 방법
- 상관계수
    - 공분산을 표준화하는 방법
    - 피어슨 상관계수, 스피어만 상관계수, 겐달 순위 상관계수
    - 상관계수 -1 <= r <= 1  # 상관계수의 값이 1에 가까우면 두 변수 사이가 끈끈하다
    
                    공분산(x,y)
상관분석(x,y) = -----------------------
                x표준편차 * y표준편차
                
x = [184,170,180]
y = [85,70,82]

import numpy as np

np.mean(x)
np.mean(y)
np.var(x) # 모집단 분산
np.var(y)
np.std(x) # 모집단 표준편차
np.std(y)

(((184-178) * (85-79)) + ((170-178)*(70-79)) + ((180-178)*(82-79))) /2

np.cov(x,y)[0][1] # 공분산
       x표본분산, xy의 공분산
array([[52., 57.],
       [57., 63.]])
  xy의공분산   y표본분산
((((184-178) * (85-79)) + ((170-178)*(70-79)) + ((180-178)*(82-79))) /2) / (df['x'].std() * df['y'].std())

np.corrcoef(x,y)[0][1] # 상관계수

import pandas as pd
df = pd.concat([pd.DataFrame(x),pd.DataFrame(y)],axis=1)
df.columns = ['x','y']
df['x'].cov(df['y']) # 공분산 
df['x'].corr(df['y']) # 상관계수

df['x'].var() # 표본분산
df['x'].std() # 표준편차
df['y'].std()

※ np.var(), np.std 와  df['x'].var,df['x'].std()는 다르다. 
앞은 모집단의 분산과 표준편차, 뒤는 표본의 분산과 표준편차 

★ 회귀분석(regression)
- 인과관계를 분석하는 방법
  어떤 변수가 어떤 변수에게 어떤 영향을 주는지를 판단(원인, 결과)
- 주어진 데이터의 독립변수(원인)로 종속변수(결과)를 예측

독립변수, 설명변수, 입력변수
- 영향을 주는 변수(원인)

종속변수, 목표변수, 결과변수
- 영향을 받는 변수(결과)

# 상관관계는 방향성(영향x) 인과관계는 영향을 주는 관계

인과관계의 조건
1. x가 변화될 때 y도 변한다.
    교육연수(독립변수) -> 생활만족도(종속변수)
2. 시간적으로 선행되어야한다.
    독립변수가 종속변수보다는 시간적으로 앞선다.
3. 외생변수를 통제해야한다.(다른 요인들을 통제하고 인과관계를 분석) 교육연수 -> 생활만족도
    다른요인(성별,나이,)

독립변수(원인) 종속변수(결과)
담배량   ->    폐암
배기량   ->    연료소비량
광고비   ->    매출액
매출액   ->    광고비
수학     ->    인공지능 전문가
영어     ->    인공지능 전문가


회귀분석을 할 때 
1. 산점도를 그려보는게 좋다.
    x가 커지면 y도 커진다.(선형)
    x가 작아지면 y는 커진다.(선형)
    x가 커지면서 y가 커지다가 작아진다.(비선형)

2. 모델의 선을 그려본다. 추세선을 예측하는 것이 회귀분석의 목적
    - OLS(Orinary Least Square)최소제곱법을 이용해서 선을 그린다.(오차의 제곱의 합을 최소로 만드는 방법)
    - 이 직선은 평균을 지난다.(평균으로 회귀)
    y = ax + b
    y : 종속변수(결과)
    x : 독립변수(원인)
    a : 회귀계수(기울기),slope
    b : 절편(y의 시작점), x가 0일 때 y값, constant
    
            y의 증가분
기울기 = --------------
            x의 증가분
            
            공분산(x,y)
a(기울기) = -------------------
            x의 분산   
            
b(절편) = 평균(y) - a * 평균(x)

height = [176,172,182,160,163,165,168,163,182,182]
weight = [72,70,70,43,48,54,51,52,73,88]

height 185일때 몸무게는?

weight = a * height + b

df = pd.DataFrame({'height':height,'weight':weight})

a = df['height'].cov(df['weight']) / df['height'].var()
b = df['weight'].mean() -a * df['height'].mean()

w = a*185 + b
w

from scipy import stats # 단순 회귀만 사용가능(다중회귀는 불가능)
slope, intercept, r_value,p_value, stderr= stats.linregress(height,weight) # 기울기, y절편 ,pvalue(유의미한지 확인하는 값 0.05 보다 작으면 기울기,절편의 값이 의미가 있는 값이다, 0.05보다 크면 의미가 없다.), 표본의 표준오차(작을수록 좋음), intercept에 대한 표준오차 등을 구해주는 함수

print('기울기',slope)
print('절편',intercept)

import matplotlib.pyplot as plt
plt.scatter(df['height'],df['weight'])
plt.plot(df['height'],df['height']*slope+intercept,c='red')


[문제205] IQ 130 일때 시험성적 예측해주세요.
import pandas as pd
from pandas import DataFrame,Series

score = pd.read_csv('c:/data/score.txt')
score.info()

score['IQ']
score['성적']

# 독립변수가 한개일 경우 단순회귀분석
slope, intercept, r_value,p_value,stderr = stats.linregress(score['IQ'],score['성적'])
slope*130 + intercept
81.988

plt.scatter(score['IQ'],score['성적'])
plt.plot(score['IQ'],score['IQ']*slope+intercept,c='red')

p_value값이 0.05보다 작으면 slope값은 의미가 있는 값이다.

# 라이브러리를 이용하여 예측
from sklearn.linear_model import LinearRegression #회귀분석 라이브러리

lr = LinearRegression()
lr.fit(score['IQ'].values.reshape(-1,1),score['성적'])
score['IQ'].values.shape # .values를 사용하면 array형식으로 바뀌어서 출력됨 Series형을 array형으로 바꿀 때 유용
score['IQ'].values.reshape(-1,1).shape # 학습데이터는 2차원배열로 되어있어야 한다.(fit할 때)
print('기울기 ',lr.coef_)
print('절편',lr.intercept_)

lr.predict([[130]])


y = a1x1 + a2x2 + a3x3 + b
# 독립변수가 2개 이상일 때는 다중회귀분석이라고 한다.
mul_lr = LinearRegression()
mul_lr.fit(score.iloc[:,2:],score['성적'])
print('기울기',mul_lr.coef_)
print('절편',mul_lr.intercept_)
mul_lr.predict([[120,3,5,2]])[0]

[문제206] 보험청구액을 예측해주세요.
insu = pd.read_csv('c:/data/insurance.csv')
# gender,smoker -> one-hot-encoding으로 바꿔줘야함 (그냥 0과 1로 바꾸면 안됨)

# one hot encoding
pd.get_dummies(insu['gender'],prefix='gender')
pd.get_dummies(insu['smoker'],prefix='smoker')

'''
from sklearn.preprocessing import OneHotEncoder
onehot_en = OneHotEncoder()
gender_onehot = onehot_en.fit_transform(insu[['gender']])
gender_onehot.toarray()
df = DataFrame(gender_onehot.toarray().astype(int))
df.columns = ['남','여']
'''

df = pd.concat([insu,pd.get_dummies(insu['gender']),pd.get_dummies(insu['smoker'])],axis=1)
df.info()

insu_df = df[['age','bmi','children','female','male','no','yes','charges']]
insu_df

lr = LinearRegression()
lr.fit(insu_df.iloc[:,0:-1],insu_df.charges)
lr.coef_
lr.intercept_
lr.predict([[10,30,3,1,0,1,0]])

'''
from collections import Counter
Counter(df.region)
'''

# ------강사님
# one hot encoding
dummy_gender = pd.get_dummies(insu['gender'],prefix='gender')

dummy_smoker = pd.get_dummies(insu['smoker'],prefix='smoker')

# 인덱스 끼리 연결하는 방법(조인)
data1 = dummy_gender.join(dummy_smoker) 
data2 = data1.join(insu[['age','bmi','children','charges']])
data2

col = ['age','bmi','children','gender_male','smoker_yes']
data2[col]
data2['charges']

lr = LinearRegression()
lr.fit(data2[col],data2['charges'])
print('기울기',lr.coef_)
print('절편',lr.intercept_)

# gender_male을 넣으면 gender_female이 기준이된다. 따라서 
gender_male의 보험청구금액이 gender_female보다  -128.63985357(기울기) 작다.
smoker_yes의 보험청구금액이 smoker_no보다 23823.39253065(기울기) 높다

# one hot encoding에서 나온 컬럼의 n-1만큼 컬럼을 넣어야한다. -> 넣지 않은 컬럼이 기준이 됨 
# 위에서는 gender_male을 넣고 gender_female이 기준이됨, smoker_yes 넣고 smoker_no가 기준이됨
# 컬럼이 3개면 2개는 넣어야함 : onehotencoding에서 나온 컬럼 n-1개를 넣고 수행해야함

col = ['age','bmi','children','gender_female','smoker_no']
data2[col]
data2['charges']

lr = LinearRegression()
lr.fit(data2[col],data2['charges'])
print('기울기',lr.coef_)
print('절편',lr.intercept_)

# gender_female을 넣으면 gender_male이 기준이된다. 따라서 
gender_female의 보험청구금액이 gender_male보다  128.63985357(기울기) 높다.
smoker_no의 보험청구금액이 smoker_yes보다 -23823.39253065(기울기) 낮다

# 학습데이터와 테스트데이터 나눠서 해보는 작업
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(data2[col],data2['charges'],test_size=0.2)

lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_train,y_train)
lr.score(x_test,y_test)

from statsmodels.formula.api import ols
ols_lr = ols('charges ~ age + bmi + children + gender_male + smoker_yes',data=data2).fit()
ols_lr.summary()
# p>|t| 값이 0.05 이하이면 의미 있는 값 , R-squared 값이 높으면 의미가 있고 좋음





```
